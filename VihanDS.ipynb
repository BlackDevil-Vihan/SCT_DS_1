```python
import pandas as pd
import matplotlib.pyplot as plt

data={"Age Group":["0-20","21-64","65+"],
     "Population":[507,807,98]
    }

df=pd.DataFrame(data)

#Barchart
plt.figure(figsize=(7,5))
plt.bar(df["Age Group"],df["Population"])
plt.title("India's Popilation Distribution by Age in 2022")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.ylim(bottom=0)
plt.show()

#Histogram

import numpy as np
ages=np.concatenate([
    np.random.randint(0,21,512),
    np.random.randint(21,65,807),
    np.random.randint(65,100,98)
])

plt.figure(figsize=(7,5))
plt.hist(ages,bins=20)
plt.title("India's Popilation Distribution by Age in 2022")
plt.xlabel("Age Group")
plt.ylabel("Population")
plt.ylim(bottom=0)
plt.show()
```
